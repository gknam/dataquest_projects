# 1. Introduction

The goal of this project is to **identify the two best markets to advertise a website in**.

The website provides courses on coding. It focuses on on mobile and web development although it covers some other areas such as data science, game development, etc. The courses are taught in English, and are accessible via monthly subscription of `$59`.



# 2. Data from existing survey

One way to collect data is to conduct surveys. However, it takes time and money. So, a practical alternative is to use existing survey results.

One such survey is [freeCodeCamp](https://www.freecodecamp.org/)'s [2017 New Coder Survey](https://medium.freecodecamp.org/we-asked-20-000-people-who-they-are-and-how-theyre-learning-to-code-fff5d668969). freeCodeCamp is a website where users learn web programming for free. The survey was participated in by more than 20,000 coders who have mostly 5 years' or less experience.

This project will use the [cleaned up data](https://github.com/freeCodeCamp/2017-new-coder-survey/blob/master/clean-data/2017-fCC-New-Coders-Survey-Data.csv) containing responses from 18,175 coders.

```python
from IPython.display import display
import pandas as pd

dtype={'CodeEventOther': 'object',
       'JobInterestOther': 'object'}

df = pd.read_csv("2017-fCC-New-Coders-Survey-Data.csv", dtype=dtype)

pd.options.display.max_columns = df.shape[1]
df.head()
```

## 2.1. Further clean the data

For string variables, I will remove leading and traling white spaces and convert characters to lower cases. This is non-essential, but makes analyses more efficient.

```python
text_cols = df.select_dtypes("object").columns
for col in df[text_cols]:
    
    # Remove leading and trailing white spaces
    df[col] = df[col].str.strip().values
    
    # Convert characters to lower cases
    df[col] = df[col].str.lower().values
```

# 3. Sample seems representative of target population

The population of interest is all coders who might use the website to learn coding. The sample (i.e. coders who participated in survey) seems to represent this population fairly well considering the following two aspects.

## 3.1. There are more new coders than experienced ones

This aspect seems to represent our target population well for two reasons. First, new coders generally need more training although coding requires life-long training. Second, compared to experienced coders, new coders are probably more interested in web-based learning compared to other means such as books. This is because interactive learning via websites makes learning much easier, and the ease will appeal more to new coders than the experienced.

```python
# Number of months spent in programming
df["MonthsProgramming"].value_counts().sort_index()
```

## 3.2. Most coders with job interest want to go into the areas which the website specializes in

When the sample is reduced to only those who specified their area of job interest (n=6992), around 83% are interested in having jobs in the areas which the website specializes in (mobile and web developments) - 5,781 coders are interested in web development.

I will use this this reduced sample from this point on.

```python
# Include only the coders who specified their area of job interest
df = df[df["JobRoleInterest"].notna()]

job_wish = df["JobRoleInterest"]

mobile = job_wish.str.contains("mobile")
web = job_wish.str.contains("web")
both = (mobile & web)

print("Number of all coders who specified job interests")
print(df.shape[0])
print()

print("Number of coders who want to learn coding for ...")
print()

print("1. Web developemnt")
print(web.sum())
print()

print("2. Mobile developemnt")
print(mobile.sum())
print()

print("3. Both")
print(both.sum())
```

# 4. Identify markets with greatest potential for profit

## 4.1. Define market

Each country can be considered a market because business prospects often vary across countries.

## 4.2. Find good markets

A good market for business is, of course, one where the business have good potential to make profit. I will try to identify them step by step.

### 4.2.1. Market with many coders

One aspect of a good market is its size. Therefore, we want to find the countries where most coders live in. Following are 10 countries where most coders reside.

```python
top_countries = df["CountryLive"].value_counts()
print(top_countries.apply(lambda x: "{} ({:.2f}%)".format(x, x / top_countries.sum() * 100)).head(10))
print()

print("Note: This information is missing for {} coders.".format(df["CountryLive"].isna().sum()))
```

The result indicates that the US and India are the two best markets to advertise in.

However, there is more to consider.

### 4.2.2. Market in which English is spoken

First, I want to target English speaking countries because the website is in English. The top 4 countries happen to meet this condition, so I'll include only them for further exploration.

```python
# Select top 4 English speaking countries
df = df[df["CountryLive"].isin(['united states of america', 'india', 'united kingdom', 'canada'])]

# Refine names
df["CountryLive"] = df["CountryLive"].replace({'united states of america': "US",
                                               'india': "India",
                                               'united kingdom': "UK",
                                               'canada': "Canada"})
```

### 4.2.3. Market with high consumer spending

Second, the profit will be potentially greater when customers have bigger budgets. For this reason, I will check, for each country, mean monthly expense for learning to code. The unit is US dollar. The monthly expenses will be checked intead of annual or weekly expenses because the website sells monthly subscriptions.

Note that I will first replace 0 with 1 in `MonthsProgramming` variable (number of months spent in programming) to avoid zero division error.

```python
# Replace 0 with 1
df["MonthsProgramming"] = df["MonthsProgramming"].replace(0, 1)

# Get each coder's monthly expense for learning to code
df["MoneyForLearningPerMonth"] = df["MoneyForLearning"] / df["MonthsProgramming"]

# Exclude coders if monthly expense is unavailable
df = df[df["MoneyForLearningPerMonth"].notna()]

def get_mean_monthly_expense(df, return_num_coders=False):

    # Get mean monthly expense for learning to code
    mean_monthly_expense = df.groupby(by="CountryLive", axis=0).agg({"MoneyForLearningPerMonth": "mean", "CountryLive": "count"}).sort_values(by="MoneyForLearningPerMonth", ascending=False)
    
    mean_monthly_expense.rename(columns={"CountryLive": "NumberOfCoders"}, inplace=True)
    
    if not return_num_coders:
        mean_monthly_expense = mean_monthly_expense[["MoneyForLearningPerMonth"]]
    
    return mean_monthly_expense

mean_monthly_expense = get_mean_monthly_expense(df)

mean_monthly_expense.head()
```

US and India came on top again. But, there is still more to consider.

#### 4.2.3.1. Remove outliers

The above result can seem unexpected because India has the lowest GDP among these countries as reported by multiple organizations. The the following table shows GDP as reported in [List of countries by GDP (PPP) per capita
](https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(PPP)_per_capita#Lists_of_countries_and_dependencies).

|&nbsp;|International Monetary Fund (2017)|World Bank (2017)|Central Intelligence Agency (1993–2017)|
|---|---|---|---|
|US|11|11|13|
|Canada|22|22|25|
|UK|26|24|30|
|India|122|119|127|

It may be that coders in India are simply willing to spend more on learning. However, it is worth considering other possibilities as well. For example, the sample might poorly represent India, Canada and United Kingdom. Or there could be outliers. I will tackle only the latter issue because it is more doable.


##### 4.2.3.1.1. Extreme outliers

```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set(style="whitegrid")

def draw_box_plot(df):
    sns.boxplot(x="CountryLive", y="MoneyForLearningPerMonth", data=df[["MoneyForLearningPerMonth", "CountryLive"]])
    plt.title("Monthly expense for coding in each country")
    plt.ylabel("Monthly expense")
    plt.xlabel("Country")
    plt.show()
    
draw_box_plot(df)
```

Obvious outliers exist for US. I will exclude coders with >= `$20000` budget.

```python
df_no_outliers = df[df["MoneyForLearningPerMonth"] < 20000]

mean_monthly_expense = get_mean_monthly_expense(df_no_outliers)
display(mean_monthly_expense.head(4))
draw_box_plot(df_no_outliers)
```

##### 4.2.3.1.2. More moderate outliers

Now that the range is reduced, some less extreme outliers are more visible. Monthly expenses appear concentrated roughly around <= `$5000` in the US, and <= `$2500` in Canada and India. So, I will reduce the dataset again accordingly.

This decision is rather intuitive because there seem to be only 2 obvious outliers (one in India and the other in Canada). However, I need to target the majority with enough budgets rather than the minority with extremely high budgets. The monthly subscription fee for the website is only `$59`, so losing a few people with > `$2500` budget will not hurt.

```python
def outliers(series, lo_lim, hi_lim):    
    lo_lim = lo_lim if lo_lim >= series.min() else series.min()
    hi_lim = hi_lim if hi_lim <= series.max() else series.max()

    is_outlier = (lo_lim > series) | (series > hi_lim)

    outlier_inds = series[is_outlier].index.tolist()
    return len(outlier_inds), outlier_inds



lims = {"India": 2500, "Canada": 3000, "US": 5000}
outlier_inds_list = []
print("Number of further outliers removed per country")
for country in lims:
    expense_per_country = df_no_outliers[df_no_outliers["CountryLive"] == country]["MoneyForLearningPerMonth"]
    num_outliers, outlier_inds = outliers(expense_per_country, 0, lims[country])
    print("{}: {}".format(country, num_outliers))
    outlier_inds_list = outlier_inds_list + outlier_inds

df_no_outliers_2 = df_no_outliers.drop(outlier_inds_list)

mean_monthly_expense = get_mean_monthly_expense(df_no_outliers_2, True)
display(mean_monthly_expense.head(4))
```

## 4.3. Best markets

After further removing 20 outliers, **US remains undisputedly the best market**. US has the highest mean monthly expense for learning as well as the highest number of coders.

The **second best market appears to be India** although this may be refutable. India seems to be better than Canada because India has more coders. However, each coder in Canada have bigger budget than those in India. A bigger budget can be helpful if, for example, a coder wants to subscribe to multiple resources. Otherwise, it won't make a difference as long as coders have enough to subscribe to the website (`$59`).

On the other hand, UK appears unsuitable because their monthly expense (`$46`) is below the monthly subscription fee for the website.

# 5. Further considerations

There are more coders in the US than in India. So, I could distribute investment accordingly: 86.5% (2918 / (2918 + 457)) of advertisement cost on US and the remaining 13.5% on India.
