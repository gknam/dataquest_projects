# 4. Draw histograms: See distribution of each column

## 4.1. Sample size

```python
# set variables for all histograms
hist_ylabel = "Frequency"
bins = 100
edgecolor = "black"


def distribution_and_summary(dataframe, columnLabel):
    # set variables for the column
    col = dataframe[columnLabel]

    # set histogram range giving leeways to min and max
    # values of the column
    col_min = col.min()
    col_max = col.max()
    col_range = col_max - col_min
    hist_range_leeway = col_range * 0.01
    hist_range = (col_min - hist_range_leeway, \
                  col_max + hist_range_leeway)

    # create figure
    fig = plt.figure(figsize=(5, 10))
    
    # draw box plot
    ax1 = fig.add_subplot(2, 1, 1)
    ax1 = col.plot(kind="box", vert=False)

    # draw histogram
    ax2 = fig.add_subplot(2, 1, 2)
    ax2 = col.hist(bins=bins, range=hist_range, edgecolor=edgecolor)

    # tweak histogram
    ax2.set_xlabel(columnLabel)
    ax2.set_ylabel(hist_ylabel)
    ax2.set_title(columnLabel + " distribution")

    # display summary description
    display(col.describe())

# draw plot
distribution_and_summary(recent_grads, "Sample_size")
```

Sample size distribution is right-skewed. Half of the majors had the sample size equal to or less than 131.

## 4.2. Median earnings

```python
distribution_and_summary(recent_grads, "Median")
```

Median earnings peaked at around mid-35K, being positively skewed.

## 4.3. Number of reported employment

```python
distribution_and_summary(recent_grads, "Employed")
```

Number of employment is also positively skewed.

## 4.4. Number of reported full time employment

```python
distribution_and_summary(recent_grads, "Full_time")
```

Another positive skew is observed here. The shape of distribution is very similar to that of the "Employed".

## 4.5. Proportion of women

```python
distribution_and_summary(recent_grads, "ShareWomen")
```

Gender distribution in each mejor seems to be almost equal.

The median is slighly greater than 0.5. Therefore, there are slighly more male-dominant majors than its counter part.

## 4.6. Unemployment rate

```python
distribution_and_summary(recent_grads, "Unemployment_rate")
```

Unemployment rate is symmetrically distribiuted in a bell shape.

## 4.7. Number of male graduates

```python
distribution_and_summary(recent_grads, "Men")
```

## 4.8. Number of female graduates
