In this project, I will create a Kaggle workflow for a machine learning competition.

The [Titanic dataset](https://www.kaggle.com/c/titanic) will be used, which comes in two sets: training (for training the model) and test (for testing the model's performance). Each row represents a passenger and each column is a feature. The target feature is the `Survived` column, and this is only in the training set. All the other features are in both datasets.

I aimed to follow the workflow shown below, but I did not go backwards after `Model selection and tuning` stage due to time constraint. Also, note that this is a modified version of the [workflow suggested by DataQuest](https://s3.amazonaws.com/dq-content/188/kaggle_workflow.svg).

![KaggleWorkflow](KaggleWorkFlow.png)

The list below summarises the tasks carried out at each stage of the workflow in this project.

* Feature processing
    * Convert data formats
        * Convert continuous features to categorical features
        * Reduce number of categories in categorical feature
    * Fill in missing values
    * Create dummy variables for each categorical feature
    * Min-max scale numerical features
    * Remove features with missing or non-numerical values
    * Add missing features filled with baseline values if training and test sets have different features.
* Feature engineering
    * Explore data
    * Create new features using existing features
* Feature selection
    * Calculate feature importance in cross-validated recursive feature elimination
    * Among the dummy variables derived from each categorical feature, remove the least important one
    * Remove features correlated together
    * Select a designated number of top features
* Model selection and tuning
    * Try different combinations of learning algorithms and hyperparameters
    * Identify the best combination
* Submission to Kaggle
    * Make prediction on test set and submit the result to Kaggle

**Note**: The terms `feature`, `variable`, and `column` will be used interchangeably in this project: `feature` will be used in general, `variable` will be used as in `dummy variable` because "dummy feature" seems to be an unusual term, and `column` will be used when features are referred to as part of a data frame.

## 2.1. Load helper functions

Following helper functions have been **created by DataQuest** for feature processing, and I modified them very slightly.

```python
# %load functions.py
def process_missing(df):
    """Handle various missing values from the data set

    Usage
    ------

    holdout = process_missing(holdout)
    """
    df["Fare"] = df["Fare"].fillna(df["Fare"].mean())
    df["Embarked"] = df["Embarked"].fillna("S") # "S" is most common value
    return df

def process_age(df):
    """Process the Age column into pre-defined 'bins' 

    Usage
    ------

    train = process_age(train)
    """
    df["Age"] = df["Age"].fillna(-0.5)
    cut_points = [-1,0,5,12,18,35,60,100]
    label_names = ["Missing","Infant","Child","Teenager","Young Adult","Adult","Senior"]
    df["Age_categories"] = pd.cut(df["Age"],cut_points,labels=label_names)
    return df

def process_fare(df):
    """Process the Fare column into pre-defined 'bins' 

    Usage
    ------

    train = process_fare(train)
    """
    cut_points = [-1,12,50,100,1000]
    label_names = ["0-12","12-50","50-100","100+"]
    df["Fare_categories"] = pd.cut(df["Fare"],cut_points,labels=label_names)
    return df

def process_cabin(df):
    """Process the Cabin column into pre-defined 'bins' 

    Usage
    ------

    train process_cabin(train)
    """
    df["Cabin_type"] = df["Cabin"].str[0]
    df["Cabin_type"] = df["Cabin_type"].fillna("Unknown")
    df = df.drop('Cabin',axis=1)
    return df

def process_titles(df):
    """Extract and categorize the title from the name column 

    Usage
    ------

    train = process_titles(train)
    """
    titles = {
        "Mr" :         "Mr",
        "Mme":         "Mrs",
        "Ms":          "Mrs",
        "Mrs" :        "Mrs",
        "Master" :     "Master",
        "Mlle":        "Miss",
        "Miss" :       "Miss",
        "Capt":        "Officer",
        "Col":         "Officer",
        "Major":       "Officer",
        "Dr":          "Officer",
        "Rev":         "Officer",
        "Jonkheer":    "Royalty",
        "Don":         "Royalty",
        "Sir" :        "Royalty",
        "Countess":    "Royalty",
        "Dona":        "Royalty",
        "Lady" :       "Royalty"
    }
    extracted_titles = df["Name"].str.extract(' ([A-Za-z]+)\.',expand=False)
    df["Title"] = extracted_titles.map(titles)
    df = df.drop("Name",axis=1)
    return df

def create_dummies(df,column_names):
    """Create Dummy Columns (Dummy Encoding) from a single Column

    Usage
    ------

    train = create_dummies(train, ["Sex", "Embarked"])
    """
    for col in column_names:
        dummies = pd.get_dummies(df[col],prefix=col)
        df = df.drop(col, axis=1)
        df = pd.concat([df,dummies],axis=1)
    return df
```

## 2.2. Impute, convert and create dummies

Missing values will be filled in, continuous features will be converted into categorical features, and dummy variables will be created from categorical features.

```python
from IPython.display import display

# Original column names
orig_cols_train = train.columns
orig_cols_holdout = holdout.columns

# Categorical column names
cat_cols = ["Pclass", "Sex", "Embarked", "Title", "Cabin_type", "Fare_categories", "Age_categories"]

# Process features
def decorator(func, *args, **kwargs):
    """source https://bit.ly/2sDASmb"""
    def decorator2(func2):
        def wrapper(x):
            return func(func2(x), *args, **kwargs)
        return wrapper
    return decorator2

@decorator(create_dummies, cat_cols)
@decorator(process_age)
@decorator(process_fare)
@decorator(process_cabin)
@decorator(process_titles)
@decorator(process_missing)
def process(df):
    return df

train = process(train)
holdout = process(holdout)

# Newly created column names (source https://stackoverflow.com/a/3428637/7194743)
new_cols_train = [col for col in train.columns if col not in orig_cols_train]
new_cols_holdout = [col for col in holdout.columns if col not in orig_cols_holdout]

# Display a sample of the newly created columns
display_df(train[new_cols_train].head())
```

## 5.2. Rescale numerical features

The numerical features are currently redundant because they have been transformed to categorical features and then to dummy variables. However, I will keep these for now to see their contribution to making predictions.

```python
from sklearn.preprocessing import minmax_scale
from sklearn.exceptions import DataConversionWarning

import warnings

# Suppress the warning "DataConversionWarning: Data with input dtype int64 was converted to float64."
warnings.filterwarnings(action="ignore", category=DataConversionWarning)

def rescale_num_cols(df):
    num_cols = ["Age", "SibSp", "Parch", "Fare"]
    for col in num_cols:
        df[col] = minmax_scale(df[col])
    return df

train = rescale_num_cols(train)
holdout = rescale_num_cols(holdout)


num_cols = ["Age", "SibSp", "Parch", "Fare"]

# Display sample of rescaled features
display(train[num_cols].head())
```

## 5.3. Remove features with missing or non-numerical values

Machine learning algorithms cannot handle these.

```python
import numpy as np
def remove_cols_with_missing_vals(df):
    miss_count = df.isna().sum()
    missing_cols = miss_count[miss_count > 0].index.tolist()
    print("Removed columns with missing values:\n{}".format(missing_cols))
    df = df.drop(missing_cols, axis=1)
    return df

def remove_non_num_cols(df):
    non_num_cols = df.select_dtypes(exclude=np.number).columns.tolist()
    print("Removed columns with non-numerical values:\n{}".format(non_num_cols))
    df = df.drop(non_num_cols, axis=1)
    return df

print("Test set")
holdout = remove_cols_with_missing_vals(holdout)
holdout = remove_non_num_cols(holdout)
print()

print("Training set")
train = remove_cols_with_missing_vals(train)
train = remove_non_num_cols(train)
```

## 5.4. Ensure training and test sets have same features

In case a feature present one set is missing in the other, the missing feature will be added.

The missing feature will be filled with baseline values (e.g. 0 if it is a dummy variable) - referred to https://stackoverflow.com/a/45365178/7194743.

```python
def fill_missing_features(train, holdout):

    train_features = set(train.drop(target, axis=1).columns)
    holdout_features = set(holdout.columns)

    missing_in_holdout = list(train_features - holdout_features)
    missing_in_train = list(holdout_features - train_features)
    
    def create_cols(df, cols, set_label):
        print("Missing columns in {} set:\n{}\n".format(set_label, cols))
        for col in cols:
            df[col] = 0
        return df
    
    if missing_in_holdout or missing_in_train:
        
        if missing_in_holdout:
            holdout = create_cols(holdout, missing_in_holdout, "holdout")
        
        if missing_in_train:
            train = create_cols(train, missing_in_train, "train")
            
        print("All missing features have been added.")
        
    return train, holdout


train, holdout = fill_missing_features(train, holdout)
```

# 6. Feature selection

## 6.1. Calculate feature importance

Importance of each feature in predicting the target value (`Survived`) will be calculated in cross-validated recursive feature elimination.

In a 10-fold cross-validation, a model will be fit with random forest classifier. This will yield importance of each feature feature in predicting the target value. Then, the least important feature will be removed and get its rank recorded.

Then, the same process will be repeated with the remaining features, each time losing the least important feature, until ranks of all features have been recorded. 

```python
from copy import deepcopy
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import KFold

def get_features_and_target(df, target):
    X = df.drop(target, axis=1)
    y = df[target]
    
    return X, y


def get_feature_importances_cv(X, y, kf):

    # Initialize list of feature importances
    feature_importances = np.zeros(X.shape[1])
    
    # For each fold
    for train_index, test_index in kf.split(X):

        # Split features (X) and target (y) into train and test sets
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]

        # Initialize and fit model
        rfc = RandomForestClassifier(random_state=random_state, n_estimators=100)
        rfc.fit(X, y)

        # Update list of feature importances
        feature_importances += rfc.feature_importances_
        
        return feature_importances
    
def get_worst_feature_ind(X, X_copy, fi):
    """
    This function is used in iterations during which 
    least important features are eliminated recursively from X_copy
    until num_final_features features are left.
    
    X is the original features.
    X_copy is a copy.
    """
    
    # Get X_copy index of worst feature
    worst_feature_iter_ind = np.where(fi == fi.min())[0][0]
    
    # Get worst feature
    worst_feature = X_copy.columns[worst_feature_iter_ind]
    
    # Get X index of worst feature
    worst_feature_all_ind = np.where(X.columns == worst_feature)[0][0]
    
    return worst_feature_all_ind, worst_feature_iter_ind

def update_feature_ranks(X, worst_feature_ind, feature_ranks, rank):
    feature_ranks[worst_feature_ind] = rank
    return feature_ranks


X, y = get_features_and_target(train, target)

# random_state is set for reproducibility
random_state = 1

# Initialize 10-fold cross validation
kf = KFold(n_splits=10, shuffle=True, random_state=random_state)

# Copy features
X_copy = deepcopy(X)

# Feature ranks according to recursive feature elimination
# based on feature importance
feature_ranks = np.ones(X.shape[1])

# Number of final (i.e. best) features to identify in recursive feature elimination
num_final_features = 1

# Repeat until num_final_features features are left
while X_copy.shape[1] > num_final_features:
    
    fi = get_feature_importances_cv(X_copy, y, kf)

    # Identify least important feature's index
    worst_feature_all_ind, worst_feature_iter_ind = get_worst_feature_ind(X, X_copy, fi)
    
    # Record least important feature's rank
    feature_ranks[worst_feature_all_ind] = X_copy.shape[1]
    
    # Remove least least important feature
    X_copy = X_copy.drop(X_copy.columns[worst_feature_iter_ind], axis=1)
    
# Display features and their importance (most important feature on top)
train_X_features = pd.Series(X.columns, name="Feature")
train_X_features_ranks = pd.Series(feature_ranks, name="Rank")
fi_all = pd.concat([train_X_features, train_X_features_ranks], axis=1).sort_values("Rank")
display(fi_all)
```

## 6.2. Remove the least important dummy variable derived from each categorical feature

One of the dummy variables for each categorical feature is redundant - this is known as the [dummy variable trap](http://www.algosome.com/articles/dummy-feature-trap-regression.html) - and therefore should be removed. That is, one dummy variable's value can be obtained from all the others' values. For example, `Sex_male` indicates either "male" or "not male". And "not male" actually indicates "female", so `Sex_female` (or `Sex_male`) can be removed. This is a good practice also because having more features makes the model more likely to be overfitted.

Instead of making a random choice, I will remove one with the least importance for each categorical feature.

```python
from pprint import pprint
import re


def remove_postfix(string):
    """
    "a_b" is changed to "a_".
    "a_b_c" is changed to "a_b_"
    """
    
    postfix = "\_(?:.(?!\_))+$" # source https://bit.ly/2OhBqJa
    return re.sub("\_(?:.(?!\_))+$", "_", string)

def remove_least_important_dummy_variable(train, holdout, fi_all, fi_dum, dum_groups, target):
    """
    train: training set (DataFrame)
    holdout: testing set (DataFrame)
    fi_dum: DataFrame with two columns
            "Feature" - Column names of dummy variables
            "Rank" - Importance of each feature (i.e. feature)
    dum_groups: Unique prefixes of each dummy variable names.
    """
    
    removed = []
    
    # For each group of dummy variables
    for col in dum_groups:

        # Get rows where feature names start with previs
        dums = fi_dum[fi_dum["Feature"].str.startswith(col)]

        # Sort the rows by feature importance
        dums = dums.sort_values(by="Rank")

        # Get the least important feature's name
        feature_to_remove = dums.iloc[[-1]]["Feature"]
        removed = removed + feature_to_remove.values.tolist()

        # Remove it from feature importances dataframe
        fi_all = fi_all[~fi_all["Feature"].isin(feature_to_remove)]
        
        # Remove it from datasets
        train = train.drop(feature_to_remove, axis=1)
        holdout = holdout.drop(feature_to_remove, axis=1)
        
        # Update features and target
        X, y = get_features_and_target(train, target)
        
    print("Following features from each dummy variable group has been removed due to having lowest feature importance:\n")
    pprint(removed)
    print("\n\n")
    
    return train, holdout, X, y, fi_all

# Get column names and feature importances of dummy variables
is_dummy = fi_all["Feature"].str.contains("_")
fi_dum = fi_all[is_dummy]

# Identify unique groups of dummy variable names
dum_groups = set(fi_dum["Feature"].apply(lambda x: remove_postfix(x)))

# Remove least important feature (feature) from each group of dummy variables
train, holdout, X, y, fi_all = remove_least_important_dummy_variable(train, holdout, fi_all, fi_dum, dum_groups, target)

print("Updated feature ranks")
display(fi_all)
```

## 6.3. Remove features if correlated together

If two features are correlated, it means that they are conveying similar information. Also, having more features increases variance of the model (i.e. model is more likely to be overfitted). Therefore, it is better to keep only one of the correlated features.

As shown below, there are no pairs of features for which pearson's correlation coefficient is over 0.9 - note that this is an arbitrary threshold. So, no feature will be removed.

```python
def get_hi_corr(X_corr):
    X_corr = X.corr()
    hi_corr_features = dict()
    for col in X_corr:
        corrs = X_corr[col]
        corrs_hi = corrs.drop(col)[abs(corrs) > 0.8]
        for ind, val in corrs_hi.iteritems():
            if (ind + " and " + col) not in hi_corr_features:
                hi_corr_features[col + " and " + ind] = val
                
                
    return hi_corr_features

hi_corr_features = get_hi_corr(X.corr())
pprint(hi_corr_features)
```

```python
def remove_feature(col_to_remove, X, train, holdout, fi_all):
    X = X.drop(col_to_remove, axis=1)
    train = train.drop(col_to_remove, axis=1)
    holdout = holdout.drop(col_to_remove, axis=1)
    fi_all = fi_all[~fi_all["Feature"].isin([col_to_remove])]
    
    return X, train, holdout, fi_all

X, train, holdout, fi_all = remove_feature("familysize", X, train, holdout, fi_all)
```

```python
def remove_less_important_feature(cols, X, train, holdout, fi_all):

    # Identify less important feature
    fi_cols = fi_all[fi_all["Feature"].isin(cols)]
    fi_cols_ranks = fi_cols["Rank"]
    is_less_important = (fi_cols_ranks == fi_cols_ranks.max())
    col_less_important = fi_cols["Feature"][is_less_important].values[0]
    
    # Remove less important feature from dataframes
    X, train, holdout, fi_all = remove_feature(col_less_important, X, train, holdout, fi_all)
    
    print("Less important and therefore removed:\n{}".format(col_less_important))
    
    return X, train, holdout, fi_all

X, train, holdout, fi_all = remove_less_important_feature(["Sex_male", "Title_Mr"], X, train, holdout, fi_all)
```

# 7. Model selection and tuning

I will now run 10-fold cross-validated searches. During the searches, 3 different models ([RandomForestClassifier()](http://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestClassifier.html), [LogisticRegression()](http://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html), [KNeighborsClassifier()](http://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html)) will be fit using various combinations of hyperparameters. Eventually, this will identify the model and parameters that yield the highest mean accuracy score.

Lists of available hyperparameters are available at each model's documentation. However, I will use those suggested by DataQuest due to time constraint.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier

def select_model(X, y, package, cv):
    
    # Track best of the best
    bob_score = 0
    bob_estimator = None
    bob_estimator_name = None
    bob_params = None
    
    # For each model and its parameters
    for i in range(len(package)):
        
        # Initialise grid search
        grid = GridSearchCV(package[i]["estimator"], \
                            package[i]["hyperparameters"], \
                            cv=cv)
        
        # Fit model
        grid.fit(X, y)
        
        # Get outputs for best estimator
        best_params = grid.best_params_
        best_score = grid.best_score_
        best_estimator = grid.best_estimator_
        
#         # Record outputs
#         package[i]["Best parameters"] = best_params
#         package[i]["Best score"] = best_score
#         package[i]["Best estimator"] = best_estimator
        
        # Display outputs
        estimator_name = package[i]["name"]
        print(estimator_name)
        print("Best parameters:", best_params)
        print("Best score:", best_score)
        print()
        
        # Select the best combination of estimator and parameters in the package
        if best_score > bob_score:
            bob_score = best_score
            bob_estimator = best_estimator
            bob_params = best_params
            bob_estimator_name = estimator_name
            
    print("Best of the best identified:")
    print("Model:", bob_estimator_name)
    print("Parameters", bob_params)
    print("Score", bob_score)
    
    return bob_estimator, bob_params



# Models (i.e. estimators) and parameters
package = \
[
    {
        "name": "KNeighborsClassifier",
        "estimator": KNeighborsClassifier(),
        "hyperparameters":
            {
                "n_neighbors": range(1,20,2),
                "weights": ["distance", "uniform"],
                "algorithm": ["ball_tree", "kd_tree", "brute"],
                "p": [1,2]
            }
    },
    {
        "name": "LogisticRegression",
        "estimator": LogisticRegression(random_state=random_state),
        "hyperparameters":
            {
                "solver": ["newton-cg", "lbfgs", "liblinear"]
            }
    },
    {
        "name": "RandomForestClassifier",
        "estimator": RandomForestClassifier(random_state=random_state),
        "hyperparameters":
            {
                "n_estimators": [4, 6, 9],
                "criterion": ["entropy", "gini"],
                "max_depth": [2, 5, 10],
                "max_features": ["log2", "sqrt"],
                "min_samples_leaf": [1, 5, 8],
                "min_samples_split": [2, 3, 5]
            }
    }
]

# Number of folds in cross validation
cv = 10


# Best of best model and parameters
bob_estimator, bob_params = select_model(X, y, package, cv)
```
